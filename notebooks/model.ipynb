# Создание подвыборки Full - из данных отправленных на раземтку

```python
one = pd.read_excel('../data/raw/Decorative/Dekorativnaja_kosmetika_multiword_15_.xlsx')
two = pd.read_excel('../data/raw/Decorative/Dekorativnaja_kosmetika_multiword_new_15.xlsx')
three = pd.read_excel('../data/raw/Decorative/Dekorativnaja_kosmetika_oneword_10_.xlsx')
four = pd.read_excel('../data/raw/Decorative/Dekorativnaja_kosmetika_oneword_new_10.xlsx')
```

```python
full_markup = pd.concat([one, two, three, four])
full_markup.reset_index(inplace=True)
```

**Получение частотности для фраз**

```python
full_m = full_markup[['Item', 'index']].merge(full, left_on='Item', right_on='item')[['item', 'frequency', 'index']]
full_m
```

### Подготовка данных

**Выбор размеченных индексов**

```python
synonyms_index = full_markup.merge(synonyms, left_on='Item', right_on='Synonyms')['index'].values
synonyms_index.shape
```

**Создание train data**

```python
negative = full_markup.loc[~full_markup['index'].isin(synonyms_index), ['Item']]
positive = full_markup.loc[full_markup['index'].isin(synonyms_index), ['Item']].sample(negative.shape[0], random_state=42)

negative['y'], positive['y'] = 0, 1
train = pd.concat([negative, positive], ignore_index=True)
train['y'].value_counts()
```

```python
import fasttext
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = fasttext.load_model('../models/pretrained/cc.ru.300.bin')
```

```python
X = [model.get_sentence_vector(str(i).replace('-', ' ').lower().strip()) for i in train['Item']]
y = train['y'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,
                                                    shuffle=True, stratify=y, random_state=42)
```

```python
from catboost import CatBoostRegressor
cat = CatBoostRegressor()
```

```python
from catboost import Pool, cv
params = {"iterations": 2000,
          "depth": 8,
          "loss_function": "RMSE",
          'eval_metric': "AUC",
          "verbose": False}
cv_dataset = Pool(data=X_train,
                  label=y_train)
scores = cv(cv_dataset,
            params,
            fold_count=2, 
            plot="True")
```
